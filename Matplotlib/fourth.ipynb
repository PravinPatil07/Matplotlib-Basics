## Bar Graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
students = np.array(['Jay','Yash','Siddharth','Devendra','Tejas','Dhammadip'])
marks = np.array([90, 85, 88, 78, 92, 80])
```

```python
plt.bar(students, marks, color='blue')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show()
```
