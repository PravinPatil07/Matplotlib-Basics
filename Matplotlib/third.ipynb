## Legend

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([1,2,3,4,5])
b = [10,20,30,40,50]
c = [17,27,37,47,57]
```

```python
fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b,'k--',label='Model length')
ax.plot(a,c,'k:',label='Data length')
ax.legend()
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

plt.show()
```

## Position Of The Legend

```python
ay = plt.subplot()
ay.plot([10,20,30],[15,16,17],'k--',label='Hello')
ay.plot([10,20,30],[25,26,27],'k:',label='World')
ay.legend(loc='lower right') # loc = 'upper right', 'upper left', 'lower left', 'lower right'
plt.show()
```

```python
ay = plt.subplot()
ay.plot([10,20,30],[15,16,17],'k--',label='Hello')
ay.plot([10,20,30],[19,17,14],'k:',label='World')
legend = ay.legend(loc='lower right') # loc = 'upper right', 'upper left', 'lower left', 'lower right'
legend.get_frame().set_facecolor('C9') # C0, C1, C2, C3, C4, C5, C6, C7, C8, C9

plt.show()
```

```python
ay = plt.subplot()
ay.plot([10,20,30],[15,16,17],'k--',label='Hello')
ay.plot([10,20,30],[19,17,14],'k:',label='World')
legend = ay.legend(loc='lower left',fontsize = 'medium') 
# loc = 'upper right', 'upper left', 'lower left', 'lower right'
#Valid font size are xx-small, x-small, small, medium, large, x-large, xx-large, larger, smaller, None
legend.get_frame().set_facecolor('C9') # C0, C1, C2, C3, C4, C5, C6, C7, C8, C9

plt.show()
```
