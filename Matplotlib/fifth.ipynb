## PyChart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
students = np.array(['Jay','Yash','Siddharth','Devendra','Tejas','Dhammadip'])
plot_in_india = np.array([90, 85, 88, 78, 92, 80])
```

```python
plt.pie(plot_in_india,labels=students,autopct='%1.2f%%') 
# autopct is used to display the percentage upto 2 decimal places
plt.title('Plot in India')

plt.show()
```
