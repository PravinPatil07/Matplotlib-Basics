## Scattered Graph

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
team_1 = np.array([87,50,46,50,90,120,20])
team_2 = np.array([30,45,90,60,80,50,60])

score_range = np.array([100,60,100,80,120,150,50])
```

```python
plt.scatter(team_1,score_range, color='r')
plt.scatter(team_2,score_range, color='b')
plt.xlabel('Teams')
plt.ylabel('Score Range')
plt.title('Scatter plot of Team 1 and Team 2')
plt.show()
```
