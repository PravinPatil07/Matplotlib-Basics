## Histograph


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
marks = np.array([10, 20,15,100, 37,30, 85,40,95,81,21,30,25,60,75,50, 60, 70, 80, 90, 100])
```

```python
plt.hist(marks, bins=[0, 20, 40, 60, 80, 100], edgecolor='black' ,color='green')
plt.title('Marks Histogram')
plt.show()
```
