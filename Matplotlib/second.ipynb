## Matplotlib ploting with pandas Dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "Score": [100, 150, 200, 250]
    ,"Balls":[60,90,150,260]
    },index=['m_1', 'm_2', 'm_3', 'm_4'])

data
```

```python
plt.plot(data["Score"],data["Balls"]) # to plot the graph
# syntax for plot is    plt.plot(x-axis,y-axis)
plt.title("Score vs Balls") # for title of the plot
plt.xlabel("Score") # for label of x-axis
plt.ylabel("Balls") # for label of y-axis
plt.grid() # for grid 
plt.show() # to show the plot
```
